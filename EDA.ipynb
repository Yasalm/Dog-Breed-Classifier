```python

# neccary libraries
from glob import glob
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import seaborn as sns
```

```python
human_files = np.array(glob("/data/lfw/*/*"))
dog_files = np.array(glob("/data/dog_images/*/*/*"))
```

```python
dog_files[99].split('/')[-2].split('.')[1]
```

```python
dogs_breed = np.array([file.split('/')[-2].split('.')[1] for file in dog_files])
    
```

```python
breed, counts = np.unique(dogs_breed, return_counts=True)
```

```python
breed[1], counts[1]
```

```python
dogs_df = pd.DataFrame({'Breed':breed, 'Counts':counts})
```

```python
dogs_df.shape
```

```python
dogs_df = dogs_df.sort_values(by='Counts', ascending=False)
dogs_df.Breed = dogs_df.Breed.str.replace('_', ' ')
```

```python
f, ax = plt.subplots(figsize=(7, 10))
# sns.set_color_codes("muted")
sns.set(style="whitegrid")
ax = sns.barplot(y='Breed', x='Counts', data=dogs_df.head(30), palette='muted', );
ax.set_xlabel('')
ax.set_ylabel('')

ax.set_title('Number of Dog Breeds - Top 30', size=15)
plt.xticks(size=12)
plt.yticks(size=13)
# ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
sns.despine(left=True, bottom=False)
```

```python
f, ax = plt.subplots(figsize=(7, 10))
# sns.set_color_codes("muted")
sns.set(style="whitegrid")
ax = sns.barplot(y='Breed', x='Counts', data=dogs_df.tail(30), palette='muted', );
ax.set_xlabel('')
ax.set_ylabel('')

ax.set_title('Number of Dog Breeds - Lowest 30', size=15)
plt.xticks(size=12)
plt.yticks(size=13)
# ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
sns.despine(left=True, bottom=False)
```
